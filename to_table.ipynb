```python
import numpy as np
import pandas as pd
import os
```

Пишем общий класс, который объединяет все файлы в нужных форматах

```python
class ConvertToObject:
    '''
    Class of all object.
    '''
    

    def to_xml(self, path_file, index = -1):
        # get path of file and open file in xml. After put in dict
        data = pd.read_xml(path_file)
        if index == -1:
            self.dict_obj[path_file] = data
        else:
            self.dict_obj[self.name_files[index]] = data

    def to_csv(self, path_file, index = -1):
        # get path of file and open file in csv. After put in dict
        data = pd.read_csv(path_file)
        if index == -1:
            self.dict_obj[path_file] = data
        else:
            self.dict_obj[self.name_files[index]] = data

    def to_xls(self, path_file, index = -1):
        # get path of file and open file in xls. After put in dict
        data = pd.read_excel(path_file)

        # убираем не нужные данные в таблице
        data = data.iloc[2:, :]
        data.drop(data.columns[[1,2,3]], axis=1 ,inplace=True)
        
        data.index = data.iloc[:, 0].tolist()
        data.drop(data.columns[[0]], axis=1 ,inplace=True)
        
        # ставим год наверх
        data.columns = data.iloc[0, :].astype(int).tolist()
        data = data.iloc[1:, :]

        if index == -1:
            self.dict_obj[path_file] = data
        else:
            self.dict_obj[self.name_files[index]] = data

        

    def __init__(self):
        self.path_files = []
        self.name_files = []
        self.dict_obj = dict()


        to_directory = os.getcwd() + "/datasets/"
        self.name_files = os.listdir(to_directory)

        for i in range(len(self.name_files)):
            self.path_files.append(os.path.join(to_directory, self.name_files[i]))
            
            end_of_namefiles = self.name_files[i][-3:]

            if end_of_namefiles == "xls":
                self.to_xls(self.path_files[i], i)
            elif end_of_namefiles == "xml":
                self.to_xml(self.path_files[i], i)
            elif end_of_namefiles == "csv":
                self.to_csv(self.path_files[i], i)



obj = ConvertToObject()
```

```python
db = obj.dict_obj.copy()
```

```python
sample_2 = db["happiness2017.csv"]
sample_2.index = sample_2.iloc[:, 0].tolist()
sample_2.drop(sample_2.columns[[0]], axis=1, inplace=True)
sample_2
```

```python
sample_3 = db["patenti.csv"]
sample_3.drop(sample_3.columns[[1,2,3,-1]], axis=1 ,inplace=True)
sample_3.index = sample_3.iloc[:, 0].tolist()
sample_3.drop(sample_3.columns[[0]], axis=1 ,inplace=True)
sample_3.columns = sample_3.columns.map(int)
sample_3
```

```python
YEAR = [2013, 2014, 2015, 2016, 2017]

GLOBAL_DATASET_YEAR = dict()

for year in YEAR:
    GLOBAL_DATASET_YEAR[year] = pd.DataFrame()
    for name in obj.name_files:
        end_of_name = name[-3:]
        if end_of_name == "xls":
            temp = obj.dict_obj[name][year].rename(name)
            GLOBAL_DATASET_YEAR[year] = pd.concat([GLOBAL_DATASET_YEAR[year], temp], axis=1)
    GLOBAL_DATASET_YEAR[year] = pd.concat([GLOBAL_DATASET_YEAR[year], sample_3[year].rename("patenti.csv")], axis=1)
    GLOBAL_DATASET_YEAR[year] = GLOBAL_DATASET_YEAR[year].join(sample_2["Happiness.Score"], how='outer')
    




```

```python
GLOBAL_DATASET_YEAR[2017]
```

```python
for year in YEAR:
    GLOBAL_DATASET_YEAR[year] = GLOBAL_DATASET_YEAR[year].loc[np.intersect1d(GLOBAL_DATASET_YEAR[year].index.tolist(), sample_2.index.tolist()), :]
```

```python
GLOBAL_DATASET_YEAR[2017]
```

```python
for year in YEAR:
    GLOBAL_DATASET_YEAR[year] = GLOBAL_DATASET_YEAR[year].fillna(GLOBAL_DATASET_YEAR[year].mean())
```

```python
GLOBAL_DATASET_YEAR[2017]
```

```python
for year in YEAR:
    GLOBAL_DATASET_YEAR[year].index.name = "COUNTRY"
    GLOBAL_DATASET_YEAR[year].drop(["Congo (Brazzaville)","Russia", "Slovakia", 
                     "Somalia", "South Korea", "Syria", 
                     "Taiwan Province of China", "Turkey", "Venezuela", 
                     "Yemen", "Congo (Kinshasa)", "Czech Republic", "Egypt",
                    "China", "Iran", "Ivory Coast", "Kyrgyzstan",
                     "Macedonia", "North Cyprus", "Palestinian Territories"], axis=0, inplace=True)
    
```

```python
for year in YEAR:
    GLOBAL_DATASET_YEAR[year]["YEAR"] = year*np.ones(len(GLOBAL_DATASET_YEAR[year]), dtype=int)
```

```python
GLOBAL_DATASET = pd.concat(GLOBAL_DATASET_YEAR.values())
```

```python
GLOBAL_DATASET = GLOBAL_DATASET.sort_values(["COUNTRY", "YEAR"])
```
